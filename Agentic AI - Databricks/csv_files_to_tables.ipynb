### Step 1: Create delta tables for csv files

```python
import os

csv_dir = "/Volumes/agentic_catalog/agentic_schema/customer_service"
csv_files = [f for f in os.listdir(csv_dir) if f.endswith('.csv')][:3]

for csv_file in csv_files:
    tbl_name = os.path.splitext(csv_file)[0]
    file_path = f"{csv_dir}/{csv_file}"
    df = spark.read.option("header", "true").csv(file_path)

    catalog_name = "agentic_catalog"
    schema_name = "agentic_schema"
    table_name = tbl_name

    df.write.format("delta").mode("overwrite").saveAsTable(f"{catalog_name}.{schema_name}.{table_name}")
    print(f"Table {catalog_name}.{schema_name}.{table_name} has been created.")
```
